### Banco de dados MNIST

O banco de dados MNIST foi construído a partir do Banco de Dados Especial 3 e do Banco de Dados Especial 1 do NIST, que contém imagens binárias de dígitos manuscritos. O NIST originalmente designou SD-3 como seu conjunto de treinamento e SD-1 como seu conjunto de teste. No entanto, o SD-3 é muito mais limpo e fácil de reconhecer do que o SD-1. A razão para isso pode ser encontrada no fato de que o SD-3 foi coletado entre os funcionários do Census Bureau, enquanto o SD-1 foi coletado entre os alunos do ensino médio. Tirar conclusões sensatas de experimentos de aprendizado requer que o resultado seja independente da escolha do conjunto de treinamento e teste entre o conjunto completo de amostras. Portanto, foi necessário construir um novo banco de dados misturando os conjuntos de dados do NIST.

O conjunto de treinamento MNIST é composto por 30.000 padrões do SD-3 e 30.000 padrões do SD-1. Nosso conjunto de teste foi composto por 5.000 padrões do SD-3 e 5.000 padrões do SD-1. O conjunto de treinamento de 60.000 padrões continha exemplos de aproximadamente 250 escritores. Garantimos que os conjuntos de escritores do conjunto de treinamento e do conjunto de teste fossem disjuntos.

SD-1 contém 58.527 imagens de dígitos escritas por 500 escritores diferentes. Em contraste com o SD-3, onde os blocos de dados de cada gravador aparecem em sequência, os dados no SD-1 são embaralhados. As identidades dos gravadores para SD-1 estão disponíveis e usamos essas informações para desembaralhar os gravadores. Em seguida, dividimos o SD-1 em dois: personagens escritos pelos primeiros 250 escritores entraram em nosso novo conjunto de treinamento. Os 250 escritores restantes foram colocados em nosso conjunto de testes. Assim, tivemos dois conjuntos com cerca de 30.000 exemplos cada. O novo conjunto de treinamento foi concluído com exemplos suficientes do SD-3, começando no padrão # 0, para fazer um conjunto completo de 60.000 padrões de treinamento. Da mesma forma, o novo conjunto de teste foi concluído com exemplos SD-3 começando no padrão # 35.000 para fazer um conjunto completo com 60.000 padrões de teste. Apenas um subconjunto de 10.000 imagens de teste (5.000 de SD-1 e 5.000 de SD-3) está disponível neste site. O conjunto de treinamento completo de 60.000 amostras está disponível.

Muitos métodos foram testados com este conjunto de treinamento e conjunto de teste. Aqui estão alguns exemplos. Detalhes sobre os métodos são fornecidos em um próximo artigo. Alguns desses experimentos usaram uma versão do banco de dados em que as imagens de entrada estavam distorcidas (computando o eixo principal da forma mais próxima da vertical e deslocando as linhas para torná-la vertical). Em alguns outros experimentos, o conjunto de treinamento foi aumentado com versões artificialmente distorcidas das amostras de treinamento originais. As distorções são combinações aleatórias de deslocamentos, dimensionamento, distorção e compressão.

## 3.1 Random Forest:

O Random Forest (floresta aleatória), como o próprio nome indica, consiste em um grande número de árvores de decisão individuais que operam como um conjunto. Cada árvore individual na floresta aleatória cospe uma previsão de classe e a classe com mais votos se torna a previsão do nosso modelo.

Um grande número de modelos relativamente não correlacionados (árvores) operando como um comitê superará qualquer um dos modelos constituintes individuais.

![](https://miro.medium.com/max/1170/1*58f1CZ8M4il0OZYg2oRN4w.png)

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
Y_prediction = random_forest.predict(X_test)
accuracy_rf=round(accuracy_score(y_test,Y_prediction)* 100, 2)
acc_random_forest = round(random_forest.score(X_train, y_train) * 100, 2)


cm = confusion_matrix(y_test, Y_prediction)
accuracy = accuracy_score(y_test,Y_prediction)
precision =precision_score(y_test, Y_prediction,average='micro')
recall =  recall_score(y_test, Y_prediction,average='micro')
f1 = f1_score(y_test,Y_prediction,average='micro')
print('Confusion matrix for Random Forest\n',cm)
print('accuracy_random_Forest : %.3f' %accuracy)
print('precision_random_Forest : %.3f' %precision)
print('recall_random_Forest : %.3f' %recall)
print('f1-score_random_Forest : %.3f' %f1)
```

```python
# plotando a matriz

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(cm, cmap=plt.cm.Blues)
ax.grid(False)
ax.set_xlabel('Valores estimados', fontsize=15, color='black')
ax.set_ylabel('Valores reais', fontsize=15, color='black')
ax.xaxis.set(ticks=range(10))
ax.yaxis.set(ticks=range(10))
ax.set_ylim(9.5, -0.5)
for i in range(10):
    for j in range(10):
        ax.text(j, i, cm[i, j], ha='center', va='center')
plt.show()
```

### 3.3 K Nearest Neighbor:

K-Nearest Neighbor é um dos algoritmos de aprendizado de máquina mais simples baseado na técnica de aprendizado supervisionado.

O algoritmo K-NN assume a semelhança entre o novo caso/dados e os casos disponíveis e coloca o novo caso na categoria que é mais semelhante às categorias disponíveis.

O algoritmo K-NN armazena todos os dados disponíveis e classifica um novo ponto de dados com base na similaridade. Isso significa que quando novos dados aparecem, eles podem ser facilmente classificados em uma categoria de conjunto de poços usando o algoritmo K-NN.

## 3.4 Gaussian Naive Bayes:

Naive Bayes é um algoritmo de classificação para problemas de classificação binária (duas classes) e multiclasse. A técnica é mais fácil de entender quando descrita usando valores de entrada binários ou categóricos.

É chamado de Bayes ingênuo ou Bayes idiota porque o cálculo das probabilidades para cada hipótese é simplificado para tornar seu cálculo tratável. Em vez de tentar calcular os valores de cada valor de atributo P(d1, d2, d3|h), eles são considerados condicionalmente independentes dado o valor alvo e calculados como P(d1|h) * P(d2|H) e em breve.

Esta é uma suposição muito forte que é muito improvável em dados reais, ou seja, que os atributos não interagem. No entanto, a abordagem tem um desempenho surpreendentemente bom em dados em que essa suposição não se sustenta.

![](https://miro.medium.com/max/1200/0*qFuHAV7Vd09064q-.jpeg)


```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
Y_pred = gaussian.predict(X_test) 
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision =precision_score(y_test, Y_pred,average='micro')
recall =  recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')
print('Confusion matrix for Naive Bayes\n',cm)
print('accuracy_Naive Bayes: %.3f' %accuracy)
print('precision_Naive Bayes: %.3f' %precision)
print('recall_Naive Bayes: %.3f' %recall)
print('f1-score_Naive Bayes : %.3f' %f1)
```

### 3.6 Decision Tree:

Uma árvore de decisão é uma estrutura semelhante a um fluxograma em que cada nó interno representa um teste em um recurso (por exemplo, se uma moeda dá cara ou coroa), cada nó folha representa um rótulo de classe (decisão tomada após computar todos os recursos) e ramos representam conjunções de recursos que levam a esses rótulos de classe. Os caminhos da raiz à folha representam regras de classificação.

![](https://miro.medium.com/max/1000/1*LMoJmXCsQlciGTEyoSN39g.jpeg)
